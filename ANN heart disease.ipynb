```python
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
hrt = pd.read_csv('heart.csv')
hrt.head()
```

```python
x1 = hrt.drop(['thal','target'],axis=1)
x1.head()
```

```python
x = hrt.drop(['thal','target'],axis=1)
x.head()
```

```python
y = hrt['target']
y.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std_scale = StandardScaler()
std_scale.fit(x1)
x_norm = std_scale.transform(x)
x_norm
```

```python
x_norm_col = pd.DataFrame(x_norm, index=x.index, columns=x.columns) 
x.update(x_norm_col)
x.head()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.4)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
print(x_train.shape[1])
print(x_train.iloc[0])
```

```python
from tensorflow.keras import models, layers

model = models.Sequential()
```

```python
model
```

```python
def build_model():
    model = models.Sequential()
    model.add(layers.Dense(20, activation = 'relu' , input_shape=(x_train.shape[1],)))
    model.add(layers.Dense(20, activation = 'relu'))
    model.add(layers.Dense(1, activation = 'sigmoid'))
    model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
    return model
```

```python
model = build_model()
```

```python
model.summary()
```

```python
hist = model.fit(x_train, y_train, epochs=10, batch_size=1, verbose=1)
```

```python
model.predict(x_test)
```

```python
model.predict_classes(x_test)
```

```python
model.evaluate(x_test,y_test)
```
